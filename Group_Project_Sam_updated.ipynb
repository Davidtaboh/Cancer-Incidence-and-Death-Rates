```python


df_state_county=df_2016["State"] + "-"+df_2016["County"]
df_2016["code"]=df_state_county
# df_2016

df_state_county_1=df_2015["State"] + "-" + df_2015["County"]
df_2015["code"]=df_state_county_1
df_2015_removed=df_2015["code"]

df_state_county_2=df_2014["State"] + "-" + df_2014["County"]
df_2014["code"]=df_state_county_2
df_2014_removed=df_2014["code"]

df_state_county_3=df_2013["State"] + "-" + df_2013["County"]
df_2013["code"]=df_state_county_3
df_2013_removed=df_2013["code"]

df_state_county_4=df_2012["State"] + "-" + df_2012["County"]
df_2012["code"]=df_state_county_4
df_2012_removed=df_2012["code"]


data_frames = [df_2016, df_2015_removed, df_2014_removed,df_2013_removed,df_2012_removed]
```
